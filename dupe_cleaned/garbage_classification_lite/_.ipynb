```python
import os
from PIL import Image

# Set the base directory where your folders are located
base_dir = '.'

# Iterate through all folders
for folder_name in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder_name)

    # Check if the current path is a directory
    if os.path.isdir(folder_path):
        # Iterate through all files in the folder
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            
            # Process only .jpg files
            if file_name.lower().endswith('.jpg'):
                try:
                    # Open the image
                    with Image.open(file_path) as img:
                        # Get original dimensions
                        width, height = img.size
                        
                        # Determine the scaling factor
                        if max(width, height) > 256:
                            if width >= height:
                                new_width = 256
                                new_height = int((height / width) * 256)
                            else:
                                new_height = 256
                                new_width = int((width / height) * 256)

                            # Resize the image
                            img_resized = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

                            # Save the resized image, overwriting the original
                            img_resized.save(file_path)
                            print(f"Resized: {file_path}")
                        else:
                            print(f"Skipped (already smaller): {file_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

print("Image resizing completed.")
```

```python
import os
import random
import shutil

# Path to the main dataset folder
dataset_dir = '.'

# Loop through all subfolders in the dataset
for folder_name in os.listdir(dataset_dir):
    folder_path = os.path.join(dataset_dir, folder_name)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        # Get a list of all image files in the subfolder
        images = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        num_images = len(images)
        
        # If there are more than 1250 images, we need to randomly delete some
        if num_images > 1250:
            # Calculate how many images to delete
            num_to_delete = num_images - 1250
            
            # Randomly select files to delete
            files_to_delete = random.sample(images, num_to_delete)
            
            for file in files_to_delete:
                file_path = os.path.join(folder_path, file)
                os.remove(file_path)  # Delete the file
                print(f"Deleted {file} from {folder_name}")

print("Dataset balancing complete.")
```
